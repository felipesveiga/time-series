<div> 
    <ul style='font-size:20px'> 
        <li> 
            A auto-correlação de uma série seria a correlação dela com sua versão de $n$ time-steps de lag.
        </li>
        <li>
            Uma série que não possui autocorrelação com uma grande quantidade de suas versões de lag é chamada de <i> white noise</i>.
        </li>
    </ul>
